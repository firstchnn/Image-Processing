```python
import cv2
import numpy as np
import math
import matplotlib.image as mpimg

%matplotlib notebook
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
bgrimg = cv2.imread('./BGR-ajred.jpg')
cv2.imwrite('./BGR-B.jpg',bgrimg[:,:,0]) 
bgr_b = cv2.imread('./BGR-B.jpg')
cv2.imwrite('./BGR-G.jpg',bgrimg[:,:,1]) 
bgr_g = cv2.imread('./BGR-G.jpg')
cv2.imwrite('./BGR-R.jpg',bgrimg[:,:,2]) 
bgr_r = cv2.imread('./BGR-R.jpg')
fig, axs = plt.subplots(2, 4)
for i in range(2):
    for j in range(4):
        axs[i-1,j-1].imshow(bgrimg)
axs[0,0].imshow(bgrimg)
axs[0,0].set_title("BGR")

axs[0,1].imshow(bgr_b)
axs[0,1].set_title("B")

axs[0,2].imshow(bgr_g)
axs[0,2].set_title("G")

axs[0,3].imshow(bgr_r)
axs[0,3].set_title("R")


cv2.imwrite('./RGB-ajred.jpg',bgrimg[:,:,[2,1,0]])
rgbimg = cv2.imread('./RGB-ajred.jpg')
cv2.imwrite('./RGB-R.jpg',rgbimg[:,:,0]) 
rgb_r = cv2.imread('./RGB-R.jpg')
cv2.imwrite('./RGB-G.jpg',rgbimg[:,:,1]) 
rgb_g = cv2.imread('./RGB-G.jpg')
cv2.imwrite('./RGB-B.jpg',rgbimg[:,:,2]) 
rgb_b = cv2.imread('./RGB-B.jpg')
axs[1,0].imshow(rgbimg)
axs[1,0].set_title("RGB")

axs[1,1].imshow(rgb_r)
axs[1,1].set_title("R")

axs[1,2].imshow(rgb_g)
axs[1,2].set_title("G")

axs[1,3].imshow(rgb_b)
axs[1,3].set_title("B")

plt.tight_layout()
plt.show()

```

```python
import math
fig, axs = plt.subplots(1, 2)
unquanimg = cv2.imread('./BGR-ajred.jpg',0)
quanimg = unquanimg
level = 4
width = quanimg.shape[0]
height = quanimg.shape[1]
for i in range (height):
    for j in range (width):
        quanimg[i,j] = math.floor((quanimg[i,j]/255)*level)
cv2.imwrite('output.jpg', quanimg)
image = cv2.imread('./BGR-ajred.jpg',0)
axs[0].set_title('image')
axs[0].imshow(image,cmap="gray")
output = cv2.imread('./output.jpg',0)
axs[1].set_title('quantized image')
axs[1].imshow(output,cmap="gray")
plt.tight_layout()
plt.show()
```

```python
data = cv2.imread('./BGR-ajred.jpg',0)
data = cv2.resize(data,(100,100))

xx, yy = np.mgrid[0:data.shape[0], 0:data.shape[1]]

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(xx, yy, data ,rstride=1, cstride=1, cmap="gray",linewidth=0)

plt.show()
```
