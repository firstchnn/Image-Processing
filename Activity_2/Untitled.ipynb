```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('./img1.jpg')
img2 = cv2.imread('./img2.jpg')
fig, axs = plt.subplots(1, 2)
axs[0].imshow(img1)
axs[1].imshow(img2)
plt.tight_layout()
plt.show()
```

```python
# cv2.VideoWriter_fourcd()
frame_size = (1920,1080)
output = cv2.VideoWriter('./output.avi', cv2.VideoWriter_fourcc('M','J','P','G'), 40, frame_size)
w1,w2 = [0,1]
div = 0.005
for i in range(200):
    sumimg = cv2.addWeighted(img1, w1 + div*i, img2, w2 - div*i, 0)
    output.write(sumimg)
for i in range(200):
    sumimg = cv2.addWeighted(img1, w2 - div*i, img2, w1 + div*i, 0)
    output.write(sumimg)
```
